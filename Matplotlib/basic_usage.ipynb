```python
# Importanto as bibliotecas do matplotlib
import matplotlib as mpl
import matplotlib.pyplot as plt
# Importanto a biblioteca numpy
import numpy as np
# Importando a biblioteca pandas
import pandas as pd
```

## Plotando um exemplo simples
*   Axes: Região de plotagem
*   Axis: Eixos da região de plotagem (pode ser 2D ou 3D)

```python
#pyplot.subplots pode criar figuras com os axes (região de plotagem)
fig, ax = plt.subplots()  # cria uma figura contendo apenas uma região de plotagem
# plota os dados nos eixos
ax.plot([1, 2, 3, 4], [1, 4, 2, 3]); # a primeira lista representa o eixo(axis) x e a segunda o eixo(axis) y 
```

```python
# definindo seed para gerador de números aleatórios
np.random.seed(42)
# Geando um dicionário com valores aleatórios para a, c e d
data = {'a': np.arange(50),
        'c': np.random.randint(0, 50, 50), # gera 50 número aleátorios entre 0 e 49
        'd': np.random.randn(50)}
data['b'] = data['a'] + 10 * np.random.randn(50)
data['d'] = np.abs(data['d']) * 100

# Gera um figura com uma região de plotagem (ax) definindo o tamanho e o layout da figura
fig, ax = plt.subplots(figsize=(5, 2.7), layout='constrained')

#scatter é um grafico de dispersão
# Os dois primeiros parâmetros são os eixos x e y, respectivamente
# O parâmetro s representa a escala (tamanho dos pontos)
# E o parâmetro c indica o array de cores
ax.scatter('a', 'b', c='c', s='d', data=data)
# Definindo o rótulo para o eixo x
ax.set_xlabel('entry a')
# Definindo o rótulo para o eixo y
ax.set_ylabel('entry b');
```

##  Estilos de Codificação
* A primeira forma de utilizar o Matplotlib é criar explicitamente as fiiguras e as regiões de plotagem e chamar métodos neles. Essa forma é orientada a objetos(OO).
* A segunda forma é deixar o pyplot automaticamente criar e gerenciar as figuras e as regiões de plotagem. E então utilizar as função funções do pyplot para plotar os gráficos 

### Estilo OO

```python
# Gera 100 valores entre 0 e 2
x = np.linspace(0, 2, 100)

# Mesmo no estilo-OO o pyplot é utilizado para gerar a figura
fig, ax = plt.subplots(figsize=(5, 2.7), layout='constrained')

# Plota um gráfico linear pois os dois primeiros parâmetros possuem valores iguais, ou seja, os valores do eixo x são iguais ao do eixo y
# Adiciona o rótulo 'linear' ao gráfico
ax.plot(x, x, label='linear')

# Plota um gráfico com os valores de x no eixo x e os valores de x elevado ao quadrado no eixo y
# Adiciona o rótulo 'quadratic' para esse gráfico
ax.plot(x, x**2, label='quadratic')

# Plota um gráfico com os valores de x no eixo x e os valores de x elevado ao cubo no eixo y
# Adiciona o rótulo 'quadratic' para esse gráfico
ax.plot(x, x**3, label='cubic')

# Adiciona o rótulo ao eixo x da região de plotagem (ax)
ax.set_xlabel('x label')
# Adiciona o rótulo ao eixo y da região de plotagem (ax)
ax.set_ylabel('y label')

# Adiciona um título para a região de plotagem (ax)
ax.set_title("Simple Plot")

# Adiciona uma legenda para os rótulos dos gráficos serem exibidos
ax.legend();  
```

### Estilo Pyplot
*   A região de plotagem (axes) não é gerada explicitamente.

```python
# Gera 100 valores entre 0 e 2
x = np.linspace(0, 2, 100)

# Gera uma figura a a partir do método figure
plt.figure(figsize=(5, 2.7), layout='constrained')
# Plota os gráficos na (implícita) região de plotagem 
plt.plot(x, x, label='linear')
plt.plot(x, x**2, label='quadratic')
plt.plot(x, x**3, label='cubic')

# Define o rótulo do eixo x e do eixo y
plt.xlabel('x label')
plt.ylabel('y label')

#Define o título para a região de plotagem
plt.title("Simple Plot")
# Adiciona a legenda
plt.legend(); 
```

## Criando uma função auxiliar
*   Útil para gerar o mesmo gráfico em diferente conjunto de dados

```python
# Criando a função auxiliar
def my_plotter(ax, data1, data2, param_dict):
    """
    Função auxiliar para gerar um gráfico
        * ax: região de plotagem (axes)
        * data1, data2: conjunto de dados do eixo x e do eixo y, respectivamente
        * **param_dict é um parâmetro adicional passado para o pyplot para estilzar os marcadores dos pontos
            Para outras opções de marcadores, veja: https://matplotlib.org/stable/api/markers_api.html

    """
    # Retorna o gráfico na região de plotagem indicada
    out = ax.plot(data1, data2, **param_dict)
    return out
```

```python
# Gera 4 conjunto de dados. Cada um com 100 valores aleatórios
data1, data2, data3, data4 = np.random.randn(4, 100)

# Gera uma figura e duas regiões de plotagem (ax1 e ax2)
# Os dois primeiros parâmetros do subplot representam, respectivamente, as linhas e as colunas da figura
# Portanto a figura terá uma linha e duas colunas
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(5, 2.7), layout='constrained')

# Chamando a função auxiliar para os conjunto de dados data1 (valores de x) e data2 (valores de y)
# Indica o dicionário com o tipo de marcador (parâmetro adicional) que será utilizado nos pontos
my_plotter(ax1, data1, data2, {'marker': 'x'})

# Chamando a função auxiliar para os conjunto de dados data3 (valores de x) e data4 (valores de y)
# Indica o dicionário com o tipo de marcador (parâmetro adicional) que será utilizado nos pontos
my_plotter(ax2, data3, data4, {'marker': 'o'});
```

## Estilizando Artists
*    Artists são basicamente tudo o que é visível na figura

```python
# Criando uma figura com uma região de plotagem (ax) e especificando a largura e a altura da figura
fig, ax = plt.subplots(figsize=(5, 2.7))

# Gerando 100 (tamanho do data1) números aleatórios
x = np.arange(len(data1))

# Plota o gráfico com x sendo os valores no eixo x e a soma aculativa de x sendo os valores no eixo y
# Define cor, largura e estilo da linha do gráfico
ax.plot(x, np.cumsum(data1), color='blue', linewidth=3, linestyle='--')

# Gera um novo gráfico, sendo os valores no eixo y igual a soma acumulativa dos valores de data2
# Modifica a cor e a largura da linha para o segundo gráfico
l, = ax.plot(x, np.cumsum(data2), color='orange', linewidth=2)
# Utiliza o método set_linestyle para modificar o estilo da linha do segundo gráfico
l.set_linestyle(':');
```

## Rotulando Plots

### Rótulos e Textos da Região de Plotagem

```python
# Atribui 115 à variável mu e 15 à variável sigma
mu, sigma = 115, 15
# Gera 10.000 número aleatórios e multiplica todos por sigma. Depois soma cada um deles com mu 
x = mu + sigma * np.random.randn(10000)

# Gerando uma figura com uma região de plotagem (ax), definindo a largura e altura da figura e o layout
fig, ax = plt.subplots(figsize=(5, 2.7), layout='constrained')

# O histograma dos dados
# x são os valores de entrada
# O segundo parâmetro indica que serão formadas 50 caixas de mesma largura no gráfico
# Density retorna a densidade de probabilidade
# Facecolor define a cor do gráfico ('C0': Azul; 'C1': Laranja)
# alpha define a transparência do gráfico, quanto menor o valor mais transparente (intervalo: 0-1)
n, bins, patches = ax.hist(x, 50, density=1, facecolor='C0', alpha=0.75)

# Rótulo do eixo x
ax.set_xlabel('Length [cm]')
# Rótulo do eixo y
ax.set_ylabel('Probability')
# título da região de plotagem
ax.set_title('Aardvark lengths\n (not really)')

# Adicionando uma string na região de plotagem
# Os dois primeiros parâmetros representam a posição no eixo x e no eixo y, respectivamente
# O terceiro parâmetro é a string que será mostrada
ax.text(75, .025, r'$\mu=115,\ \sigma=15$')

# Configurando os eixos de acordo com a sequencia xmin, xmax, ymin, ymax
ax.axis([55, 175, 0, 0.03])
# Habilita grid no gráfico
ax.grid(True); 
```

### Anatações
* Útil para destacar pontos ou partes importantes do gráfico 

```python
# Gerando uma figura com uma região de plotagem (ax), definindo a largura e altura da figura
fig, ax = plt.subplots(figsize=(5, 2.7))
# Gera números de 0 a 4.99 com passo de 0.01
t = np.arange(0.0, 5.0, 0.01)
# para cada valor de t gera um novo valor de acordo com a equação: cos(2*pi*t)
s = np.cos(2 * np.pi * t)

# Plot com os vaores de t no eixo x e os valores de s no eixo y
# lw representa a largura da linha
# a variável line pode ser utilizada para modificar artists do gráfico utilizando métodos específicos
line, = ax.plot(t, s, lw=2)

# Anotação na região de plotagem ax
# O primeiro parâmetro indica o texto que será mostrado
# O segundo parâmetro indica para qual ponto a anotação será feito e terceiro indica a posição do texto da anotação
# Arrowprops configura a cor (facecolor) e o tamanho (shrink) da seta
ax.annotate('local max', xy=(2, 1), xytext=(3, 1.5),
            arrowprops=dict(facecolor='black', shrink=0.05))

# Limite para o eixo y (ymin e ymax, respectivamente)
ax.set_ylim(-2, 2); 
```
