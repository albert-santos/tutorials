```python
# Importanto as bibliotecas do matplotlib
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

## Plotando um exemplo simples
*   Axes: Região de plotagem
*   Axis: Eixos da região de plotagem (pode ser 2D ou 3D)

```python
#pyplot.subplots pode criar figuras com os axes (região de plotagem)
fig, ax = plt.subplots()  # cria uma figura contendo apenas uma região de plotagem
# plota os dados nos eixos
ax.plot([1, 2, 3, 4], [1, 4, 2, 3]); # a primeira lista representa o eixo(axis) x e a segunda o eixo(axis) y 
```

## Tipos de Entradas para Plotar Funções
*   Para plotar funções é necessário utillizar entradas do tipo numpy array
*   É possível converter alguns tipos para numpy array utilizando numpy.asarray()

```python
# Convertendo uma matriz numpy para um array numpy
b = np.matrix([[1, 2], [3, 4]])
b_asarray = np.asarray(b)

# Convertendo um Pandas dataframe para numpy array
lst = ["Jay","Raj","Jack"]
names = pd.DataFrame(lst, columns = ['Name'])
print(type(names))
names_asarray = np.asarray(names)
print(type(names_asarray))
```

```python
# definindo seed para gerador de números aleatórios
np.random.seed(42)
# Geando um dicionário com valores aleatórios para a, c e d
data = {'a': np.arange(50),
        'c': np.random.randint(0, 50, 50), # gera 50 número aleátorios entre 0 e 49
        'd': np.random.randn(50)}
data['b'] = data['a'] + 10 * np.random.randn(50)
data['d'] = np.abs(data['d']) * 100

# Gera um figura com uma região de plotagem (ax) definindo o tamanho e o layout da figura
fig, ax = plt.subplots(figsize=(5, 2.7), layout='constrained')

#scatter é um grafico de dispersão
# Os dois primeiros parâmetros são os eixos x e y, respectivamente
# O parâmetro s representa a escala (tamanho dos pontos)
# E o parâmetro c indica o array de cores
ax.scatter('a', 'b', c='c', s='d', data=data)
# Definindo o rótulo para o eixo x
ax.set_xlabel('entry a')
# Definindo o rótulo para o eixo y
ax.set_ylabel('entry b');
```

##  Estilos de Codificação
* A primeira forma de utilizar o Matplotlib é criar explicitamente as fiiguras e as regiões de plotagem e chamar métodos neles. Essa forma é orientada a objetos(OO).
* A segunda forma é deixar o pyplot automaticamente criar e gerenciar as figuras e as regiões de plotagem. E então utilizar as função funções do pyplot para plotar os gráficos 

### Estilo OO

```python
# Gera 100 valores entre 0 e 2
x = np.linspace(0, 2, 100)

# Mesmo no estilo-OO o pyplot é utilizado para gerar a figura
fig, ax = plt.subplots(figsize=(5, 2.7), layout='constrained')

# Plota um gráfico linear pois os dois primeiros parâmetros possuem valores iguais, ou seja, os valores do eixo x são iguais ao do eixo y
# Adiciona o rótulo 'linear' ao gráfico
ax.plot(x, x, label='linear')

# Plota um gráfico com os valores de x no eixo x e os valores de x elevado ao quadrado no eixo y
# Adiciona o rótulo 'quadratic' para esse gráfico
ax.plot(x, x**2, label='quadratic')

# Plota um gráfico com os valores de x no eixo x e os valores de x elevado ao cubo no eixo y
# Adiciona o rótulo 'quadratic' para esse gráfico
ax.plot(x, x**3, label='cubic')

# Adiciona o rótulo ao eixo x da região de plotagem (ax)
ax.set_xlabel('x label')
# Adiciona o rótulo ao eixo y da região de plotagem (ax)
ax.set_ylabel('y label')

# Adiciona um título para a região de plotagem (ax)
ax.set_title("Simple Plot")

# Adiciona uma legenda para os rótulos dos gráficos serem exibidos
ax.legend();  
```

### Estilo Pyplot
*   A região de plotagem (axes) não é gerada explicitamente.

```python
# Gera 100 valores entre 0 e 2
x = np.linspace(0, 2, 100)

# Gera uma figura a a partir do método figure
plt.figure(figsize=(5, 2.7), layout='constrained')
# Plota os gráficos na (implícita) região de plotagem 
plt.plot(x, x, label='linear')
plt.plot(x, x**2, label='quadratic')
plt.plot(x, x**3, label='cubic')

# Define o rótulo do eixo x e do eixo y
plt.xlabel('x label')
plt.ylabel('y label')

#Define o título para a região de plotagem
plt.title("Simple Plot")
# Adiciona a legenda
plt.legend(); 
```
