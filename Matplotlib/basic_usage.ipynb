```python
# Importanto as bibliotecas do matplotlib
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

## Plotando um exemplo simples
*   Axes: Região de plotagem
*   Axis: Eixos da região de plotagem (pode ser 2D ou 3D)

```python
#pyplot.subplots pode criar figuras com os axes (região de plotagem)
fig, ax = plt.subplots()  # cria uma figura contendo apenas uma região de plotagem
# plota os dados nos eixos
ax.plot([1, 2, 3, 4], [1, 4, 2, 3]); # a primeira lista representa o eixo(axis) x e a segunda o eixo(axis) y 
```

## Tipos de Entradas para Plotar Funções
*   Para plotar funções é necessário utillizar entradas do tipo numpy array
*   É possível converter alguns tipos para numpy array utilizando numpy.asarray()

```python
# Convertendo uma matriz numpy para um array numpy
b = np.matrix([[1, 2], [3, 4]])
b_asarray = np.asarray(b)

# Convertendo um Pandas dataframe para numpy array
lst = ["Jay","Raj","Jack"]
names = pd.DataFrame(lst, columns = ['Name'])
print(type(names))
names_asarray = np.asarray(names)
print(type(names_asarray))
```

```python
# valor da seed do gerador de números aleatórios
np.random.seed(42)

# gerando um dicionário com valores aleatórios para a, c e d
data = {'a': np.arange(50),
        'c': np.random.randint(0, 50, 50), # Gera 50 números aleatórios entre 0 e 49
        'd': np.random.randn(50)}
data['b'] = data['a'] + 10 * np.random.randn(50)
data['d'] = np.abs(data['d']) * 100

# Gera uma figura e um axes (região de plotagem) definindo o tamanho da figura e do layout
fig, ax = plt.subplots(figsize=(5, 2.7), layout='constrained')

# Scatter é um gráfico de dispersão
# Os dois primeiros parâmetros do scatter representam os eixos x e y
# O parâmetro s define a escala (tamanho) dos pontos e o parâmetro c é o array de cores 
ax.scatter('a', 'b', c='c', s='d', data=data)
# Definindo o rótulo do eixo x
ax.set_xlabel('entry a')
# Definindo o rótulo do eixo y
ax.set_ylabel('entry b');
```
